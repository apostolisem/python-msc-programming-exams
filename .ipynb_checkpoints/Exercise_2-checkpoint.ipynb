```python
# References
# https://stackoverflow.com/questions/45489846/remove-rows-from-pandas-dataframe-based-on-condition
# https://stackoverflow.com/questions/38243556/how-to-apply-function-to-multiple-pandas-dataframe
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html
# https://pypi.org/project/pycountry-convert/
# https://en.wikipedia.org/wiki/ISO_3166-1_alpha-2
# https://www.php.net/manual/en/function.geoip-continent-code-by-name.php (for the continent codes)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pycountry_convert import country_alpha3_to_country_alpha2, country_alpha2_to_continent_code

# load the CSV files in a list of datasets
corporate_profit = pd.read_csv('oecd_corporate_profit.csv')
tax_income = pd.read_csv('oecd_tax_income.csv')
tax_on_property = pd.read_csv('oecd_tax_on_property.csv')
unemployment = pd.read_csv('oecd_unemployment.csv')
df = {'corporate_profit': corporate_profit, 'tax_income': tax_income, 'tax_on_property': tax_on_property, 'unemployment': unemployment}

#### Helper Functions

# get continent name by country
def get_continent(countrycode):
    '''
    get the continent for a country code in Alpha3 format
    @param countrycode string
    
    return string
    '''
    continents = {'AF':'Africa','AN':'Antarctica','AS':'Asia','EU':'Europe','NA':'North America','OC':'Oceania','SA':'South America'}    
    try:
        alpha2 = country_alpha3_to_country_alpha2(countrycode)
        cont = country_alpha2_to_continent_code(alpha2)        
        cont = continents[cont]
    except:
        cont = 'Unknown'
    return cont

# get 5 randomly selected countries
def get_random_countries_list(df):
    '''
    get the country list for a df
    @param df dataframe
    
    return list
    '''
    countries_list = df.LOCATION.unique()    
    return countries_list[np.random.randint(0,len(countries_list)-1, 5)]


# remove the OAVG, EU averages
def remove_OAVG_EU_avg(df):
    OAVG_EU_country_list = ["OAVG", "EU28", "OECD","EA19","EU27_2020","G-7"] # list of OAVG, EU averages, found them by analyzing the unique values of LOCATION
    df = df[~df['LOCATION'].isin(OAVG_EU_country_list)].reset_index(drop=True)
    return df
df = {k: v.pipe(remove_OAVG_EU_avg) for k, v in df.items()}

# drop columns FLAG_CODES
def drop_FLAG_CODES(df):
    df = df.drop(columns=['Flag Codes'])
    return df
df = {k: v.pipe(drop_FLAG_CODES) for k, v in df.items()}
```

```python
# plot for 5 countries the tax on personal income across all years
tax_income_df = df['tax_income'].copy()
countries_list = get_random_countries_list(tax_income_df) # get 5 countries to plot
for country in countries_list: # loop through 5 randomly chosen countries
    new_df = pd.DataFrame(tax_income_df[tax_income_df['LOCATION'] == country]).reset_index(drop=True).set_index('TIME', drop=True) # new df for current country
    new_df['Value'].plot(title='Income Tax for '+country, xticks=new_df.index, rot=45)  # generate a new plot
    plt.show()
```

```python
# plot the distributions of the tax on personal income for the full dataset
tax_income_df['Value'].plot(kind='hist', bins=20, title='Tax Distribution') # plot a histogram for the full dataset
plt.show()
```

```python
# plot the above per continent
tax_income_df['CONTINENT'] = tax_income_df['LOCATION'].apply(get_continent) # add a column with the corresponding continent
tax_income_df_group = tax_income_df.groupby('CONTINENT') # group by continent
for group_name, df_group in tax_income_df_group: # plot tax distribution per continent
    df_group['Value'].plot(kind='hist', bins=20, title='Tax Distribution for '+group_name)
    plt.show()
```

```python
# find the average and median tax on corporate profits per country
corporate_profit_df = df['corporate_profit'].copy()
corporate_profit_df['CONTINENT'] = corporate_profit_df['LOCATION'].apply(get_continent) # add a column with the corresponding continent
corporate_profit_df_groupped = corporate_profit_df.groupby('LOCATION')
corporate_median = corporate_profit_df_groupped['Value'].agg(np.median).reset_index() # mediam
corporate_avg = corporate_profit_df_groupped['Value'].agg(np.average).reset_index() # average
corporate_stats = corporate_median.merge(corporate_avg, on='LOCATION', how='right').rename(columns={"Value_x": "Median", "Value_y": "Average"})
print(corporate_stats)
```

```python
# calculate the ratio of the tax on personal income versus the tax on property for each country. 
# Then plot across the years for few selected countries (5).
tax_income_per_country = df['tax_income'][['LOCATION','TIME','Value']].groupby('LOCATION')
tax_property_per_country = df['tax_on_property'][['LOCATION','TIME','Value']].groupby('LOCATION')
countries_list = get_random_countries_list(tax_income_per_country) # get 5 countries to plot
for group_name, df_group in tax_income_per_country:
    income_df = df_group.rename(columns={"Value": "Income Tax"})
    property_df = tax_property_per_country.get_group(group_name).rename(columns={"Value": "Property Tax"}).drop(columns='LOCATION')
    income_vs_property_df = income_df.merge(property_df, on='TIME', how='right')
    if group_name in countries_list:
        income_vs_property_df.plot(title='Income Tax vs Property Tax for '+group_name, y=['Income Tax','Property Tax'], x='TIME', xticks=income_vs_property_df.TIME, rot=45)  # generate a new plot
        plt.show()
```

```python
# calculate the unemployment rate per country per year.
unemployment_df = df['unemployment'].copy() # copy the original df
unemployment_df_per_country = unemployment_df.groupby('LOCATION') # group
new_unemployment_df_per_country = [] # a new dataset
for group_name, df_group in unemployment_df_per_country:
    df_group['TIME'] = df_group['TIME'].apply(lambda x: x[0:4]) # remove the month from TIME
    df_group = df_group.groupby(['LOCATION','INDICATOR','SUBJECT','MEASURE','FREQUENCY','TIME'], as_index=False).agg({'Value': 'mean'}) # average per country
    new_unemployment_df_per_country.append(df_group) 
new_unemployment_df_per_country = pd.concat(new_unemployment_df_per_country).reset_index(drop=True) # combine all new groups together
print(new_unemployment_df_per_country)
```

```python
# Do the same per continent.
unemployment_df = df['unemployment'].copy() # copy the original df
unemployment_df['CONTINENT'] = unemployment_df['LOCATION'].apply(get_continent) # add a column with the corresponding continent
unemployment_df_per_continent = unemployment_df.groupby('CONTINENT')
new_unemployment_df_per_continent = [] # create a new empty dataset
for group_name, df_group in unemployment_df_per_continent:
    df_group['TIME'] = df_group['TIME'].apply(lambda x: x[0:4]) # remove the month from TIME
    df_group = df_group.groupby(['CONTINENT','INDICATOR','SUBJECT','MEASURE','FREQUENCY','TIME'], as_index=False).agg({'Value': 'mean'}) # average value per CONTINENT
    new_unemployment_df_per_continent.append(df_group) # append to the new dataset
new_unemployment_df_per_continent = pd.concat(new_unemployment_df_per_continent).reset_index(drop=True) # combine all groups together
print(new_unemployment_df_per_continent)
```

```python
# Standardize the unemployment rate values per continent.
new_unemployment_df_per_continent['Standardized'] = new_unemployment_df_per_continent.groupby('CONTINENT')['Value'].transform(lambda x: (x-x.mean())/ x.std() if pd.api.types.is_numeric_dtype(x) else x) # normalize the rates
print(new_unemployment_df_per_continent)
```

```python
# Rank the countries in decreasing order of the average unemployment rate for the years 2011-2015.
unemployment_df = df['unemployment'].copy() # copy the original df
unemployment_df['TIME'] = unemployment_df['TIME'].apply(lambda x: x[0:4]) # remove the month from TIME column
unemployment_df['TIME'] = unemployment_df['TIME'].apply(pd.to_numeric) # turn TIME column to numeric
unemployment_df = unemployment_df.loc[(unemployment_df['TIME'] >= 2011) & (unemployment_df['TIME'] <= 2015)] # keep only records for 2011 to 2015
unemployment_df = unemployment_df.groupby(['LOCATION','INDICATOR','SUBJECT','MEASURE','FREQUENCY'], as_index=False).agg({'Value': 'mean'}) # average the values for years 2011-15
unemployment_df['TIME'] = '2011-2015' # add a column time for the corresponding years
unemployment_df = unemployment_df.sort_values(by='Value', ascending=False) # sort in decreasing order
print(unemployment_df) 
```
